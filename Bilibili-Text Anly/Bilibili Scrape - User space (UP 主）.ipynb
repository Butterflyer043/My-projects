```python
# import library
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import datetime
import time 
# 
```

```python
# =============================================================================
# Reference :
# https://www.bilibili.com/read/cv3430609/  -- Bilibili api
# https://github.com/31415926535x/CollegeProjectBackup/blob/master/BilibiliSpiders
# =============================================================================

videosCount = 1 # number of video 
videosID = 0
flag = True

# mid = up主的UID 
mid = str(20165629) 
videos = []
# pn = page number
pn = 1 
# ps = the number of videos in each page,  max=100
ps = 30 
if(ps > 100):
    ps = 100
ps=str(ps) 
# 
df = pd.DataFrame(columns=['title', 'aid', 'bvid','time','length','play','description'])

while(videosID < videosCount):
    url = "https://api.bilibili.com/x/space/arc/search?mid=" + mid + "&ps=" + '20' + "&tid=0&pn=" + str(pn) + "&keyword=&order=pubdate&jsonp=jsonp"
    response = requests.get(url)
    
    # js_loads return a dict: 
        # dict_keys(['code', 'message', 'ttl', 'data'])
    js_loads = json.loads(response.text)
    
    # data is a dict: dict_keys(['list', 'page'])
    data = js_loads['data']
    
    
    
    if(flag):
        # data : dict_keys(['list', 'page'])
        # data["page"] = {'count': 2436, 'pn': 122, 'ps': 20}
        # count = total videos number
        videosCount = data['page']['count']
        flag = False
    pn += 1 

    
    # data['list'] = dict_keys(['tlist', 'vlist'])
    # tlist: (dict)number of videos for each category 
    # vlist: (list of dicts) video info 
    vlist = data['list']['vlist']
    # count how many videos in one page
    vlistCount = len(vlist)
    
    for i in range(0, vlistCount):
        video = vlist[i]
        
        df=df.append({'title':video['title'],'aid':video['aid'],'bvid':video['bvid'],'typeid':video['typeid'],
                      'comment':video['comment'],'time':video['created'],'length':video['length'], 
                      'play':video['play'], 'description':video['description']},ignore_index=True)


        videosID += 1
        # print(str(videosID) + " av" + str(aid))
        # v = (videosID, aid, pic, des, title, length)
        #v = (videosID, aid)
        #videos.append(v)


```

```python
# convert timestamp to readable data
df['date'] = pd.to_datetime(df['time'],unit='s')
df=df[['title','description','date','aid','bvid','time','length','play','typeid','comment']]
df.head()
```

```python
df.to_csv("共青团视频全列表.csv")
```

```python
# Only extract the video between "20/11/2019 00:00:00" to "20/03/2020 23:59:59"
# string = "20/01/2020 00:00:00"
string1 = "20/11/2019 00:00:00"
string2 = "20/03/2020 23:59:59"
# convert date to timestamp
begin_date = time.mktime(datetime.datetime.strptime(string1, "%d/%m/%Y %H:%M:%S").timetuple())
end_date = time.mktime(datetime.datetime.strptime(string2, "%d/%m/%Y %H:%M:%S").timetuple())
print("Begin date-timestamp:{} \nEnd date-timestamp:{}".format(begin_date,end_date)) 
```

```python
df_video = df[(df['time']>= begin_date) & (df['time']<=end_date)]
```

```python
df_info = pd.DataFrame(columns=['aid', 'view', 'danmaku', 'reply', 'favorite', 'coin', 'share', 'now_rank', 
         'his_rank', 'like', 'dislike', 'no_reprint', 'copyright'])
for i,row in df_video.iterrows():
    aid = row['aid']
    url = 'http://api.bilibili.com/archive_stat/stat?aid='+str(aid)
    response = requests.get(url)
    # get video infomation
    vinfo = json.loads(response.text)
    info=vinfo['data']
    df_info = df_info.append(info, ignore_index=True)
df_info.head()
```

```python
info
```

```python
# df_full = video + info (between period)

# combine two dataframe by aid
df_full = pd.merge(df_video, df_info, on='aid')
# remove duplicated cols
df_full.head(10)
```

```python
df_type = pd.read_csv('typeid.csv')
```

```python
df_type = df_type.drop('Unnamed: 0',axis=1)
df_type
```

```python
# df_full2  = full + type info
df_full2 = pd.merge(df_full,df_type,on='typeid')
df_full2 = df_full2.sort_values(by='time')
df_full2 
```

```python
df_full2=df_full2.reset_index(drop=True)
```
