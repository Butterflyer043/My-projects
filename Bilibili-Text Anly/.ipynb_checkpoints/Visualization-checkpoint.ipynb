# Videos Contents Analysis  
 
Has the video released by the Communist Youth League changed before and after the lockdown period of Wuhan?  
共青团视频内容分析: 共青团发布的视频是否在武汉封城前后有变化。 
### Load data
载入数据

```python
# import library
import pandas as pd
import datetime
import time 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data
df = pd.read_csv('共青团info_cleaned.csv')
df = df.drop('Unnamed: 0',axis=1)
df.head(1)
```

```python
df.shape
```

```python
df.loc[df.typeid == 59,"section"] = '音乐'
df.loc[df.typeid == 59,"sub-section"] = '原创音乐'
```

```python
t1 = "18/01/2020 00:00:00"
t2 = "08/04/2020 00:00:00"
begin_date = time.mktime(datetime.datetime.strptime(t1, "%d/%m/%Y %H:%M:%S").timetuple())
end_date = time.mktime(datetime.datetime.strptime(t2, "%d/%m/%Y %H:%M:%S").timetuple())

#  videos released before the Wuhan lockdown
df_before = df[(df['time']<=begin_date)]
#  videos released during the Wuhan lockdown
df_lock = df[(df['time']>= begin_date) & (df['time']<=end_date)]
#  videos released after the Wuhan lockdown
df_after = df[(df['time']>= end_date)]
```

```python
# display the sanb 
df.section.unique()
#df.loc[df.section=="影视",:]
```

```python
df['sub-section'].unique()
```

## By Video types

Comparing the released video type difference between the before and after the lockdown period of Wuhan.   

对比武汉封城前后的发布的视频类型

```python
# pie chart video type 
# sub section 
from pylab import *
import matplotlib.colors as colors
```

```python
df_pie = pd.DataFrame(
{"labels" : list(df.section.unique()),
"before" :np.nan ,
"lock" :np.nan,
"after" : np.nan},)
#test_dict2 = {test_keys2[i]: test_values2[i] for i in range(len(test_keys2))} 

for i in list(df.section.unique()) :
    try:
        df_pie.loc[df_pie.labels == i,"before"] = int(pd.DataFrame(df_before.groupby(['section']).count()["aid"]).loc[i])
    except:
        pass

for i in list(df.section.unique()) :
    try:
        df_pie.loc[df_pie.labels == i,"lock"] = int(pd.DataFrame(df_lock.groupby(['section']).count()["aid"]).loc[i])
    except:
        pass
    
for i in list(df.section.unique()) :
    try:
        df_pie.loc[df_pie.labels == i,"after"] = int(pd.DataFrame(df_after.groupby(['section']).count()["aid"]).loc[i])
    except:
        pass
    
df_pie
```

### By Category
按B站大分区

```python

# plot the graph in Chinese
plt.figure(figsize=(20,10))

# set up color
#cmap = plt.get_cmap("viridis",256)
cmap = plt.cm.get_cmap('Set3', 10)(np.arange(10))
#cmap = colors.ListedColormap(cmap(np.arange(10)))
# set up figure 
the_grid = GridSpec(1, 3)

# plot df_before
for i in range(3):
    state = ["before","lock","after"][i]
    tit = ['封城前','封城时','封城后'][i]
    plt.subplot(the_grid[0, i], aspect=1, title=tit)
    names=df_pie.loc[:,["labels",state]].dropna(thresh=2)["labels"]
    size=df_pie.loc[:,["labels",state]].dropna(thresh=2)[state]
    ind = df_pie.loc[:,["labels",state]].dropna(thresh=2).index
    plt.rcParams["font.family"] = 'Arial Unicode MS'
    plt.rcParams['axes.unicode_minus'] = False
    plt.pie(size, labels=names, colors=cmap[ind],autopct='%1.1f%%' )
    #, textprops={'color':"w"} 
plt.show()

```

```python
df_pie["labels2"] = ['News', 'Technology', 'Music', 'Lifestyles',
                     'Fashion', 'Documentary', 'Animation', 
                     'Film and TV', 'Animation(Chinese)', 'Game']

# plot the graph in Chinese
plt.figure(figsize=(20,10))

# set up color
#cmap = plt.get_cmap("viridis",256)
cmap = plt.cm.get_cmap('Set3', 10)(np.arange(10))
#cmap = colors.ListedColormap(cmap(np.arange(10)))
# set up figure 
the_grid = GridSpec(1, 3)

# plot df_before
for i in range(3):
    state = ["before","lock","after"][i]
    tit = ['Before Lockdown','During Lockdown','After Lockdown'][i]
    plt.subplot(the_grid[0, i], aspect=1, title=tit)
    names=df_pie.loc[:,["labels2",state]].dropna(thresh=2)["labels2"]
    size=df_pie.loc[:,["labels2",state]].dropna(thresh=2)[state]
    ind = df_pie.loc[:,["labels2",state]].dropna(thresh=2).index
    #plt.rcParams["font.family"] = 'Arial Unicode MS'
   # plt.rcParams['axes.unicode_minus'] = False
    plt.pie(size, labels=names, colors=cmap[ind],autopct='%1.1f%%' )
    #, textprops={'color':"w"} 
plt.show()
```

### By Subcategory

```python
df_bar = pd.DataFrame(
{"labels" : list(df_before.loc[df_before.section=="资讯",:].groupby(['sub-section']).count()['aid'].keys()),
"before" :np.nan ,
"lock" :np.nan,
"after" : np.nan},)

# news section
for n in range(3):
    d = [df_before,df_lock,df_after][n]
    t = ["before","lock","after"][n]
    for i in list(d.loc[d.section=="资讯",:].groupby(['sub-section']).count()['aid'].keys()) :
        try:
            df_bar.loc[df_bar.labels == i,t] = int(pd.DataFrame(d.loc[d.section=="资讯",:].groupby(['sub-section']).count()["aid"]).loc[i])
        except:
            pass

# tech section 
df_bar2 = pd.DataFrame(
{"labels" : list(df.loc[df.section=="科技",:].groupby(['sub-section']).count()['aid'].keys()),
"before" :np.nan ,
"lock" :np.nan,
"after" : np.nan},)
for n in range(3):
    d = [df_before,df_lock,df_after][n]
    t = ["before","lock","after"][n]
    for i in list(d.loc[d.section=="科技",:].groupby(['sub-section']).count()['aid'].keys()) :
        try:
            df_bar2.loc[df_bar2.labels == i,t] = int(pd.DataFrame(d.loc[d.section=="科技",:].groupby(['sub-section']).count()["aid"]).loc[i])
        except:
            pass
df_bar2["lock"] = df_bar2["lock"].fillna(0)
df_bar2

# other
df_bar3 = pd.DataFrame(
{"labels" : ["原创音乐-音乐","短片-影视","日常-生活","人文历史-纪录片"],
"before" :np.nan ,
"lock" :np.nan,
"after" : np.nan},)
for n in range(3):
    d = [df_before,df_lock,df_after][n]
    t = ["before","lock","after"][n]
    for i in ["原创音乐-音乐","短片-影视","日常-生活","人文历史-纪录片"]:
        try:
            df_bar3.loc[df_bar3.labels == i,t] = int(pd.DataFrame(d.groupby(['typename']).count()["aid"]).loc[i])
        except:
            pass
#df_bar3.["before"] = df_bar3["before"].fillna(0)
df_bar3.iloc[1] = df_bar3.iloc[1].fillna(0)
new_row = df_bar3.iloc[1] + df_bar3.iloc[3] # 人文历史-纪录片 plus 短片-影视
new_row.name = 4
df_bar3=df_bar3.append([new_row])
#new_row.name = 'Short Film & Documentary'
df_bar3
```

```python
plotdata = pd.DataFrame({
    df_bar.labels[0]:list(df_bar.iloc[0,1:4]),
    df_bar.labels[1]:list(df_bar.iloc[1,1:4]),
    df_bar.labels[2]:list(df_bar.iloc[2,1:4]),
    df_bar.labels[3]:list(df_bar.iloc[3,1:4]),
    }, 
    index=["封城前", "封城中", "封城后"]
)
plotdata2 = pd.DataFrame({
    df_bar2.labels[0]:list(df_bar2.iloc[0,1:4]),
    df_bar2.labels[1]:list(df_bar2.iloc[1,1:4]),
    df_bar2.labels[2]:list(df_bar2.iloc[2,1:4]),
    df_bar2.labels[3]:list(df_bar2.iloc[3,1:4]),
    }, 
    index=["封城前", "封城中", "封城后"]
)

plotdata3 = pd.DataFrame({
    "原创音乐-音乐":list(df_bar3.iloc[0,1:4]),
    "日常-生活":list(df_bar3.iloc[2,1:4]),
    "短片-影视,人文-纪录片":list(df_bar3.iloc[4,1:4]),
    }, 
    index=["封城前", "封城中", "封城后"]
)

# plot the graph in Chinese

plt.rcParams["font.family"] = 'Arial Unicode MS'
fig, (ax1, ax2,ax3) = plt.subplots(1, 3,figsize=(20,6))

plotdata.plot.bar( ax=ax1, title='资讯类视频子分类',color=plt.cm.get_cmap('tab20b', 5)(np.arange(4)))
ax1.set(xlabel='时期', ylabel='视频数')
ax1.set_xticklabels(["封城前", "封城中", "封城后"],rotation = 0)

plotdata2.plot.bar( ax=ax2,title='科技类视频子分类',color=plt.cm.get_cmap('tab20b', 5)(np.arange(4)))
ax2.set(xlabel='时期', ylabel='视频数')
ax2.set_xticklabels(["封城前", "封城中", "封城后"], rotation = 0)

plotdata3.plot.bar( ax=ax3,title='其他高发布的子分类',color=plt.cm.get_cmap('tab20b', 5)(np.arange(4)))
ax3.set(xlabel='时期', ylabel='视频数')
ax3.set_xticklabels(["封城前", "封城中", "封城后"], rotation = 0)

```
