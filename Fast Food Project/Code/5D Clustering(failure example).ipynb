## Format all-item dataset

```python
import pandas as pd
import numpy as np
import re
df = pd.read_csv("all_item.csv")
df = df.drop(["Unnamed: 0"],axis = 1)

### format the restaurant name for future use
name_list = df.Restaurant.unique()
for i in np.arange(len(name_list)):
    name_list[i] = re.sub("[^0-9A-Za-z ]","",name_list[i]).lower()

new_restname = []
for name in df.Restaurant:
    name = re.sub("[^0-9A-Za-z]","",name).lower()
    new_restname.append(name)
df['Restaurant'] = new_restname
```

## Clustering Salad & Sides dataset

```python
import numpy as np
from sklearn.cluster import KMeans 
```

```python
df_salad_sides = df_salad_sides.reset_index(drop = True)
df_salad_sides.dropna(axis=0, how='any', inplace=True)
X = df_salad_sides[['Calories','Protein (g)','Carbs (g)','Sodium (mg)','Sugars (g)']]
```

```python
Result=KMeans(n_clusters=3, verbose=1)
Result.fit(X)
```

```python
labels = Result.labels_
centroids = Result.cluster_centers_
print(labels)
print(centroids)
```

```python
healthy_level = []
for i in range(len(labels)):
    if labels[i] == 0:       
        healthy_level.append('healthy')
    elif labels[i] == 1:
        healthy_level.append('medium')
    elif labels[i] == 2:
        healthy_level.append('unhealthy')
```

```python
df_salad_sides['Healthy_level'] = healthy_level
```
