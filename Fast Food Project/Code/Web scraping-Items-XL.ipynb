```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

pd.set_option('display.max_columns', None)

def scrape (restaurant):
        
    """
    scrape nutrition table and return row_list and column name list
    
    """

    url = "https://www.nutrition-charts.com/"  + restaurant
    response = requests.get(url)
    response.status_code 
    soup = BeautifulSoup(response.content, 'html.parser')
    stat_table = soup.find_all('table' )  # get table 
    stat_table = stat_table[0]
    
    row_list = []
    col_name = []
    
    for row in stat_table.find_all('tr'):
        temp = []
        temp2 = []
        for cell in row.find_all('th'):
            temp2.append(cell.text) 
        col_name.append(temp2) # get column name 
        for cell in row.find_all('td'):
            temp.append(cell.text) 
        row_list.append(temp) # get list of each row 
    return row_list, col_name
    
       
# =============================================================================
# reformat  dataframe
# =============================================================================
#### clean col_name
def colunm_name(col_name ):
    col_name = [x for x in col_name if x != []] # remove empty list
    col_name = col_name[0]
    col_name[0] = 'Items' # rename column - "Items"
    col_name.insert(0,"Category") # rename  - "category"
    return col_name

#### clean row_list
# clean sub table's title 
def category_title (row_list):
    row_list = [x for x in row_list if x != []]
    ind_item=[] # lndex list of sub-table's title
    for ind, val in enumerate(row_list): 
        if len(val) == 1:
            ind_item.append(ind)
            val_temp = re.sub('back to top', '', val[0]) # remove unecessary string  
            row_list[ind][0] = re.sub('\n', '', val_temp )   # remove unecessary string  
           # print(val)
    # add category title for each row
    for i in range(len(ind_item)):
        try:
            a = ind_item[i]
            #print(row_list[ind_item[i]])
            b = ind_item[i+1]
            title = row_list[ind_item[i]][0]
            #print("a=",a,"b=",b,"title=",title)
            for r in range(a+1,b):
                row_list[r].insert(0,title)
                #print(r)
        except IndexError:
            a = ind_item[i]
            b = len(row_list)
            title = row_list[ind_item[i]][0]
            #print("oh,""a=",a,"b=",b,"title=",title)
            for r in range(a+1,b):
                row_list[r].insert(0,title)
                
                
    row_list = [x for x in row_list if len(x) != 1]
        #row_list.insert(0, category_title)
    return row_list


def reformat_type (df):  
    df.Category = df.Category.astype ('category') 
    df.Items =  df.Items.astype ('category')  
    
    for i in range(2,7):
        df.iloc[:,i] = pd.to_numeric (df.iloc[:,i],errors='coerce')
    df = df.dropna()
    df.iloc[:,2:7] = df.iloc[:,2:7].astype ('int32') 
    return df
    
```

```python
#######  Fast Food Restaurant Rank    
#######  https://www.esquire.com/food-drink/restaurants/a15120/biggest-fast-food-11215111/

# =============================================================================
# Mcdonalds
# =============================================================================
#https://www.nutrition-charts.com/mcdonalds
restaurant =  "Mcdonalds"
row_list, col_name = scrape (restaurant)
col_name = colunm_name(col_name)
col_name.pop()
row_list =   category_title (row_list)     
df = pd.DataFrame(row_list,columns=col_name)
### Data wrangling
df.dtypes
df = reformat_type (df)
#df.Category.cat.categories.tolist()
#df.Items.cat.categories.tolist()
### Data wrangling
df.to_csv("Restaurant 01 - "+restaurant+'.csv') #export to csv
```
