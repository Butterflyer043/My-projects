## Clustering Salad & Sides dataset

```python
import numpy as np
from sklearn.cluster import KMeans 
```

```python
df_salad_sides = df_salad_sides.reset_index(drop = True)
df_salad_sides.dropna(axis=0, how='any', inplace=True)
X = df_salad_sides[['Protein (g)','Carbs (g)','Sugars (g)']]
```

```python
Result=KMeans(n_clusters=3, verbose=1)
Result.fit(X)
labels = Result.labels_
centroids = Result.cluster_centers_
```

```python
df_salad_sides['labels'] = labels
cluster0 = df_salad_sides[df_salad_sides['labels'] == 0]
cluster1 = df_salad_sides[df_salad_sides['labels'] == 1]
cluster2 = df_salad_sides[df_salad_sides['labels'] == 2]
```

```python
fig=plt.figure(figsize = (8,6))
ax=Axes3D(fig)
ax.scatter(cluster0["Protein (g)"], cluster0["Carbs (g)"], cluster0["Sugars (g)"],color='orange')
ax.scatter(cluster1["Protein (g)"], cluster1["Carbs (g)"], cluster1["Sugars (g)"],color='green')
ax.scatter(cluster2["Protein (g)"], cluster2["Carbs (g)"], cluster2["Sugars (g)"],color='blue')
centers = np.array(Result.cluster_centers_)
ax.scatter(centers[:,0], centers[:,1], centers[:,2], marker="X", color='red')
ax.set_xlabel('Protein')
ax.set_ylabel('Carbs')
ax.set_zlabel('Sugars')
plt.show()
```

```python
fig.savefig('Salad and Sides Clustering.jpg')
```

We can figure out that the orange points has the smallest protein, carbs and sugars so we consider those items with orange color as healthy, then green points as medium and blue points as unhealthy. 

```python
healthy_level = []
for i in range(len(labels)):
    if labels[i] == 0:       
        healthy_level.append('healthy')
    elif labels[i] == 1:
        healthy_level.append('medium')
    elif labels[i] == 2:
        healthy_level.append('unhealthy')
```

```python
df_salad_sides['Healthy_level'] = healthy_level
```

## Recommendation Function

```python
def get_restaurant(city):
    cityname = city.lower()
    target = rest[rest['city'] == cityname]
    restname = target.name.unique()
    return restname        
```

```python
def beverage_recom(health,restname):
    ret_bever = {}
    total_item = []
    for restaurant in restname:
        temp = df_beverage[df_beverage['Restaurant'] == restaurant]    
        df_ret_item = temp[temp['Healthy_level'] == health]
        tempitem = []
        for items in df_ret_item.Items:
            tempitem.append(items)
        ret_bever[restaurant] = tempitem
        total_item += tempitem
        total = len(total_item)
    return ret_bever, total_item, total
    
```

```python
def salad_sides_recom(health,restname):
    ret_salad_sides = {}
    total_item = []
    for restaurant in restname:
        temp = df_salad_sides[df_salad_sides['Restaurant'] == restaurant]    
        df_ret_item = temp[temp['Healthy_level'] == health]
        tempitem = []
        for items in df_ret_item.Items:
            tempitem.append(items)
        ret_salad_sides[restaurant] = tempitem
        total_item += tempitem
        total = len(total_item)
    return ret_salad_sides, total_item, total
```

```python
def entree_recom(health,restname):
    ret_entree = {}
    total_item = []
    for restaurant in restname:
        temp = df_entree[df_entree['Restaurant'] == restaurant]    
        df_ret_item = temp[temp['Healthy_level'] == health]
        tempitem = []
        for items in df_ret_item.Items:
            tempitem.append(items)
        ret_entree[restaurant] = tempitem
        total_item += tempitem
        total = len(total_item)
    return ret_entree, total_item, total
```

```python
def get_recommendation(city,bever_health,sides_health,entree_health):
    
    city = city.lower()
    
    
    rest = get_restaurant(city)
    
    bever_dic, bever_item, bever_count = beverage_recom(bever_health,rest)
    sides_dic, sides_item, sides_count = salad_sides_recom(sides_health,rest)
    entree_dic, entree_item, entree_count = entree_recom(entree_health,rest)
    
    bever_recomm = []
    for key in bever_dic:
        if  len(bever_dic[key]) == 0:
            continue
        temp = str(bever_dic[key][0]) + ' from ' + str(key)
        bever_recomm.append(temp)
    
    sides_recomm = []
    for key in sides_dic:
        if  len(sides_dic[key]) == 0:
            continue
        temp = str(sides_dic[key][0]) + ' from ' + str(key)
        sides_recomm.append(temp)
        
    entree_recomm = []
    for key in entree_dic:
        if  len(entree_dic[key]) == 0:
            continue
        temp = str(entree_dic[key][0]) + ' from ' + str(key)
        entree_recomm.append(temp)
        
    print('In '+ city +' We recommend you the following dishes:\n')
    
    print('For ' + bever_health + ' beverages, we recommend:\n')
    
    for item in bever_recomm:
        print(item + '\n')
        
    print('For ' + sides_health + ' salads or sides, we recommend:\n')
        
    for item in sides_recomm:
        print(item + '\n')
          
    print('For ' + entree_health + ' entree, we recommend:\n')
          
    for item in sides_recomm:
        print(item + '\n')      
      
```

```python
def main():
    
    cityname = pd.read_csv('Restaurant.csv')
    df_beverage = pd.read_csv('beverage.csv')
    df_salad_sides = pd.read_csv('salad_and_sides.csv')
    df_entree = pd.read_csv('entree.csv')
    
    while True:
        city = input('Enter your city name\n')
        city = city.lower()
        
        if city not in list(cityname.city):
            print('Sorry, city not in list, please choose another city.')
            main()
            
    
        bever_health = input('Please enter healthy level for beverage(healthy, medium or unhealthy)\n')
        sides_health = input('Please enter healthy level for salads or sides(healthy, medium or unhealthy)\n')
        entree_health = input('Please enter healthy level for entree(healthy, medium or unhealthy)\n')
    

        get_recommendation(city,bever_health,sides_health,entree_health)
    
        
    
        
```
